### Methodology 
**Mean**- Mean is the average of the data

**Median**- The middlemost value of the dataset that divides the data into two equal halves.

**Variance**- Variance measure the dispersion of the data from the mean
 It is an average of the sum of squares of difference between an observation and the mean. Thus the variance is always positive.The variance is based on all the observations
 
**Standard deviation**- Standard deviation is the positive square root of the variance.
Like variance, it also measures the variability of the observations from the mean

**Quartiles**- Quartiles are the data points which partitions the data into 4 equal parts. So, there are 3 quartiles.

*The First Quartile*: The first quartile is the point that partitions the first 25% of the data.This means that 25% of the data are less than or equal to the first quartile and 75% of the observations are greater than or equal to the first quartile. The first quartile is the same as the 25th percentile.

*The Second Quartile*: The second quartile is the point that partitions the data in two equal parts. This means that 50% of the data are less than or equal to the second quartile and 50% of the observations are greater than or equal to the second quartile. The second quartile is the same as the 50th percentile or median.

*The Third Quartile*: The third quartile is the point that partitions the data in 75-25 ratio. This means that 75% of the data are less than or equal to the third quartile and 25% of the observations are greater than or equal to the third quartile. The third quartile is called the 75th percentile.

**Mode**-Mode of the data is the value which has the highest frequency

**Skewness**-Skewness helps us to study the shape of the data.
It represents how much a distribution differs from a normal distribution, either to the left or to the right.
The value of the skewness can be either positive, negative or zero

**Kurtosis**-Kurtosis measures the peakedness of the distribution
In other words, kurtosis is a statistical measure that defines how the tails of the distribution differ from the normal distribution
Kurtosis identifies whether the tails of a given distribution contain extreme values

## CRV-Central Limit Theorem Verification
 
Verify the Central Limit Theorem by taking random samples from the simulated data and calculating sample means. You can observe how
the sample means approximate a normal distribution and understand the importance of the
CLT in statistics.

```python
# define a function to obtain the sample means for given population
# pass the population and and number of samples as the input parameters
def sample_mean_calculator(data, n_samples):
    
    # create an empty list to store the means of each sample
    sample_means = []
    for i in range(n_samples):
       
 # generates a random sample of size 500 from a given array or a list
        sample = np.random.choice(data, size=500, replace=False)
        
        # calculate the mean of the sample data
        sample_mean = np.mean(sample)
        
        # append the mean to the list
        sample_means.append(sample_mean)    
        
    return sample_means
```

## CRV-Probability Calculations
Calculate probabilities related to the continuous (such as normal) distribution. For example, you can find the probability that a randomly selected value
falls within a specified range or above/below a certain threshold.

Let us find P(-0.2 <= X <= 0.1)

```python
# standardize the variable with x = -0.2
# z = x - mu / sigma 
z_2 = ((-.2 - mean) / std)
# 'cdf()' returns the P(Z <= z) i.e P(Z <= -0.2)
p_2= stats.norm.cdf(z_2)
# standardize the variable with x = 0.1
z_1 = (0.1 - mean) / std

# 'cdf()' returns the P(Z <= z) i.e P(Z <= 0.1)
p_1 = stats.norm.cdf(z_1)

# calculate the difference between 'p_1' and 'p_2' to find the required probability
prob = p_1 - p_2

# use 'round()' to round-off the value to 4 digits
req_prob = prob
```

```python
print('The probability of X lying between -0.2 to 0.1 is: ', round(req_prob,4))
```

### Transition Matrix Simulation

```python
trans_mat = np.array([
[0.6, 0.3, 0.1],
[0.5, 0.2, 0.3],
[0.7, 0.1, 0.2]
])
# initial state probabilities
initial_probabilities = np.array([0.2, 0.5, 0.3])
# Number of steps in the simulation
num_steps = 15
# Simulate the Markov chain
current_state = np.random.choice(range(len(initial_probabilities)),p=initial_probabilities)
for step in range(num_steps):
    print(f"Step {step + 1}: State {current_state + 1}")
    current_state = np.random.choice(range(len(trans_mat)),p=trans_mat[current_state])
# Calculate state probabilities after the specified number of steps
# initial_state_prob (I) * transition_matrix (T)^(num_steps[10])
# I.T^10
final_probabilities = np.dot(initial_probabilities, np.linalg.matrix_power(trans_mat, num_steps))
# Print the final state probabilities
print(f"\nState Probabilities after {num_steps} steps:")
for i, prob in enumerate(final_probabilities):
    print(f"State {i + 1}: {prob:.4f}")
```

In the above code, we are randomly selecting the next state using random.choice function with it's respective probability and assigning it to the next state. 

Now after 15 steps , the final probabilities we get for State 1 , State 2 and State 3 are 0.5922,0.2427,0.1650 respectively. These probabilities are merely based on the previous probabilities of the previous states assigned.

### Recurrent Events
Recurrent events in a Markov chain refer to the situation where the system repeatedly returns to certain states.

Let's consider a simple example where we model the states of a system with three recurrent events: event 0,event 1 and event 2. 

The Markov chain can transition between these events based on specified transition probabilities.

```python
import numpy as np

class RecurrentEvent:
    def __init__(self, transition_matrix):
        self.transition_matrix = transition_matrix
        self.number_states = len(transition_matrix)
        self.state = 0

    def event(self):
        next_state = np.random.choice(range(self.number_states), p=self.transition_matrix[self.state])
        self.state = next_state
transition_matrix = np.array([[0.3, 0.7, 0.0],
                              [0.4, 0.3, 0.3],
                              [0.2, 0.3, 0.5]])

recurrentmodel = RecurrentEvent(transition_matrix)

# Simulate multiple events
num_simulations = 15
for _ in range(num_simulations):
    recurrentmodel.event()
    print("Current State:", recurrentmodel.state)
```

In the above output we can see the sequence of states visited. Initially the it toggles between state 1 and state 0 and after about 6 steps it stays at state 2 for another 2 steps. Again we can see that for the next 5 steps, the states oscillate between 0 and 1.

### Ergodicity

```python
#Ergodicity
import numpy as np

transition_matrix = np.array([[0.3, 0.7, 0.0],
                              [0.4, 0.3, 0.3],
                              [0.2, 0.3, 0.5]])

n_states = 3

# Initial state
current_state = np.random.choice(range(n_states))

n_steps = 500
sequences = []
state_prob = np.zeros(n_states)

# Simulate Markov chain
for i in range(n_steps):
    sequences.append(current_state)
    current_state = np.random.choice(range(n_states), p=transition_matrix[current_state])

for sequence in sequences:
    state_prob[sequence] += 1

state_prob /= n_steps  # This is the time-averaged probability

# Calculate stationary distribution (also known as equilibrium distribution)
time_prob = np.linalg.matrix_power(transition_matrix, 500)[0]

# Display results
print("State Probabilities:", state_prob)
print("Time averaged probabilities:", time_prob)
```

In the above output we see  that the time-averaged probabilities are pretty much close to the state probabilities by which we can understand that the over a period of time the time spent in each state eventually converges to stead state.

# 4.1.4 Variance Reduction Techniques

```python
def before_sampling_distribution(x):
    return np.exp(-x**2) / np.sqrt(np.pi)

num_samples = 10000
samples = np.random.normal(0, 1, num_samples)

calculated_integral = np.mean(before_sampling_distribution(samples))
print("Estimated Integral before applying Importance Sampling:", calculated_integral)
```

```python
import numpy as np

def before_sampling_distribution(x):
    return np.exp(-x**2) / np.sqrt(np.pi)

def importance_sampling(x):
    return np.exp(-x**2) / (2 * np.sqrt(np.pi))

num_samples = 10000
samples = np.random.normal(0, 1, num_samples)
weights = before_sampling_distribution(samples) / importance_sampling(samples)

estimated_integral = np.mean(weights * samples)
print("Estimated Integral after Importance Sampling:", estimated_integral)
```

In the above example, I am taking normally distributed sample of 10000 data points. Now before applying any reduction technique , we can see that the integral value calculated comes out to be  0.0326. But after we do some importance sampling in which we basically give more weight to the outcomes that are rare but contribute significantly to the output, we see that the result improves and our integral value becomes 0.0058

```python
import itertools
import random

# Function to perform random permutation
def random_elements(elements):
    # Generate all permutations
    all_permu = list(itertools.permutations(elements))
    
    # Randomly select a permutation
    random_permu = random.choice(all_permu)
    
    return random_permu

# Set of elements
elements = [1, 2, 3, 4]

# Perform the simulation
random_permu = random_elements(elements)

# Print the result
print("Random Permutation:", random_permu)
```

# 4.2.1 Bayes' Theorem

The theorem is generally expressed in the following formula:

$$ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} $$

#### Application in Our Analysis
To calculate the probability of a person having Parkinson's disease given the values of features like "HNR" (Harmonics-to-Noise Ratio) and "NHR" (Noise-to-Harmonics Ratio) using the simple Bayes formula, we would use Bayes' Theorem. 

The formula for Bayes' Theorem is as follows:

$$ P(Parkinson’s∣HNR,NHR)=\frac{P(HNR,NHR∣Parkinson’s)\times P(Parkinson’s)}{ P(HNR,NHR)} $$

Where:

**P(Parkinson’s∣HNR,NHR)** is the probability of having Parkinson's disease given the values of HNR and NHR.

**P(HNR,NHR∣Parkinson’s)** is the likelihood of observing the given values of HNR and NHR given that the person has Parkinson's disease

**P(Parkinson’s)** is the  probability of having Parkinson's disease

**P(HNR,NHR)** is the probability of observing the given values of HNR and NHR

```python
# Defining the thresholds
nhr_val =0.024
hnr_val = 22

# Calculating the prior probability (P(Parkinsons))
parkinsons_prob = df1['status'].mean()

# Calculating the likelihood (P(NHR > 0.024, HNR < 22 | Parkinsons))
parkinsons_patients = df1[df1['status'] == 1]
prob_nhr_hnr_given_parkinsons = ((df1['NHR'] > nhr_val) & 
                                (df1['HNR'] < hnr_val)).mean()

# Calculating the marginal likelihood (P(NHR >0.024, HNR < 22 ))
p_nhr_hnr = ((df1['NHR'] > nhr_val) & 
                 (df1['HNR'] < hnr_val)).mean()

#final probability calculation
p_parkinsons_given_nhr_hnr = (prob_nhr_hnr_given_parkinsons * parkinsons_prob) / p_nhr_hnr

print(p_parkinsons_given_nhr_hnr)

```

**Inference** : From the above results we can see 75 % of people have parkinsons disease when the NHR value is higher than 0.024 and HNR value is lower than 22.

```python
#Shapiro Wilk test 
from scipy.stats import shapiro

# Assuming df1 is your DataFrame, and you want to check the normality of each column
for column in df1.columns:
    if df1[column].dtype in ['float64', 'int64']:
        stat, p_val = shapiro(df1[column])
        print(f"Shapiro-Wilk test for '{column}': Statistic={stat:.4f}, p-value={p_val:.4f}")
```

**Inference**: We can see that a lot of columns do not have p_value > 0.05 which indicates that they are not normally distributed. We observed this pattern above as well when we saw the skewness with the help of histograms.

Let us see how would the shpiro test show results for 'spread2' column

# 4.2.3 Factor Analysis

Factor analysis is a linear statistical model. It is used to explain the variance among the observed variable and condense a set of the observed variable into the unobserved variable called factors.


Each factor explains a particular amount of variance in the observed variables. It helps in data interpretations by reducing the number of variables.


```python
# Import required libraries
import pandas as pd
from sklearn.datasets import load_iris
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
```

**Adequacy Test**
Before we perform factor analysis, we need to evaluate the “factorability” of our dataset. Factorability means "can we find the factors in the dataset?". There are two methods to check the factorability or sampling adequacy:

*Bartlett’s Test*

*Kaiser-Meyer-Olkin Test*

**Bartlett’s test of sphericity** checks whether or not the observed variables intercorrelate at all using the observed correlation matrix against the identity matrix. If the test found statistically insignificant, we cannot move forward with factor analysis.

```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_value,p_val=calculate_bartlett_sphericity(df1)
chi_value, p_val
```

In this Bartlett ’s test, the p-value is 0 which indicates that the observed correlation matrix is not an identity matrix. Thus the test is insignificant

**Kaiser-Meyer-Olkin (KMO) Test** measures the suitability of data for factor analysis.Value of KMO less than 0.6 is considered inadequate.

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_a,kmo_m=calculate_kmo(df1)
kmo_m
```

The overall KMO for our data is 0.79, which is fairly good. This value indicates that we can proceed with planned factor analysis.

#### Choosing the Number of Factors
For choosing the number of factors, we can use the Kaiser criterion and scree plot. Both are based on eigenvalues.

The eigenvalue is a good criterion for determining the number of factors. Generally, an eigenvalue greater than 1 will be considered as selection criteria for the feature.

The graphical approach is based on the visual representation of factors' eigenvalues also called scree plot. This scree plot helps us to determine the number of factors where the curve makes an elbow.

```python
from factor_analyzer.factor_analyzer import FactorAnalyzer 
fa = FactorAnalyzer(n_factors=10,rotation='varimax')
fa.fit(df1)
# Check Eigenvalues
eigv, vec = fa.get_eigenvalues()
eigv
```

```python
sorted(eigv)
```

Above we can see 5 eigenvalues are greater than 1 so 5 factors (or unobserved variables).

```python
# Create scree plot using matplotlib
plt.scatter(range(1,df1.shape[1]+1),eigv)
plt.plot(range(1,df1.shape[1]+1),eigv)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

The scree plot method draws a straight line for each factor and its eigenvalues. Number of eigenvalues greater than one are considered as the number of factors.
